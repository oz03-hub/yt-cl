```python
import pandas as pd
```

```python
df_transcripts = pd.read_csv("data/transcripts_new.csv")
df_meta = pd.read_csv('data/english_samples.csv', usecols=["id", "title", "description", "view_count", "like_count", "duration", "channel_follower_count", "categories"], encoding_errors="ignore")
df_speech_duration = pd.read_csv("speech_detection_results.csv")
df_gpt_categories = pd.read_csv("data/merged_categorized.csv", usecols=["video_id", "openai_response"])
df_audio_features = pd.read_csv("clustered_results.csv", usecols=["File", "ZCR", "SpectralCentroid", "SpectralFlux", "RMSEnergy"])
```

```python
df = df_transcripts.merge(df_meta, left_on="video_id", right_on="id", how="inner")
df = df.merge(df_speech_duration, left_on="video_id", right_on="video_id", how="inner")
df = df.merge(df_gpt_categories, left_on="video_id", right_on="video_id", how="inner")
df = df.merge(df_audio_features, left_on="video_id", right_on="File", how="inner")
df.drop(columns=["File"], inplace=True)
df.head()
```

```python
df.to_csv("data/transcripts_with_features.csv", index=False)
```
